# 📥 Cargar datasets (entry point) en un Notebook

Antes de comenzar a analizar datos, el primer paso es cargar el dataset en nuestro entorno de trabajo (Notebook en Google Colab).

En Python, esto lo hacemos generalmente utilizando la librería Pandas, que nos permite trabajar con la estructuras de datos `DataFrames` que presentamos en la Clase 2.

Existen varias formas de cargar datasets en un notebook, comencemos a ver algunas:

# 1. Cargar un dataset subido al Notebook

Primero debemos subir el archivo .csv o .xlsx al entorno del notebook (carpeta en el menú lateral) y luego podemos cargarlo fácilmente con el método read_csv de pandas: `pd.read_csv`

Nota 1: si el archivo no está previamente subido al entorno, retorna error.

Nota 2: el archivo se elimina cuando Google reinicia la Máquina Virtual sobre la que se está ejecutando el Notebook.

```python
# Importar la librería Pandas
import pandas as pd
```

```python
# pd.read_csv recibe el nombre del archivo y retorna un dataframe
df = pd.read_csv('tips.csv')
```

```python
# Visualizamos los primeros registros del dataframe
df.head(3)
```

# 2. Cargar un dataset usando el link del archivo compartido de Google Drive

Si tenemos un archivo .csv en una carpeta de nuestro Google Drive, primero compartimos sólo ese archivo en modo lectura y copiamos el link o enlace.

Luego usamos el siguiente código para importarlo como Dataframe en este Notebook

```python
# Importar la librería Pandas
import pandas as pd
```

```python
# Link generado al compartir el archivo
link = "https://drive.google.com/file/d/1DjuuUbTYIOkluqs3xM07-lYbs8bCqLpN/view?usp=drive_link"

# Usamos regex para extraer el file_id del link
file_id = link.split('/d/')[1].split('/')[0]

# Con ese file_id generamos la url
url = f"https://drive.google.com/uc?id={file_id}"

# Finalmente lo importamos usando el método read_csv() con la url generada
df = pd.read_csv(url)
df.head()
```

# 4. Cargar un dataset usando el link del Spreadsheet compartido

También podemos trabajar directamente con datos alojados en un Spreadsheet de Google Sheets, lo cual es muy útil cuando compartimos datasets entre varias personas.

Para hacerlo, necesitamos el link del archivo y asegurarnos de que esté configurado como público o con acceso de lectura.

* Paso 1: Obtener el identificador del archivo (eventualmente además el identificador de la hoja)

Si tenés un link como este:
`https://docs.google.com/spreadsheets/d/1FRLM9E7W4VVWEP_sx9QWq2w0okWfbGJ3IWFFXRL0-70/edit?gid=1937180264#gid=1937180264`

Debés transformarlo a:
`https://docs.google.com/spreadsheets/d/ID_DEL_DOC/export?format=csv`

reemplazando, en este caso, `ID_DEL_DOC` por
`1FRLM9E7W4VVWEP_sx9QWq2w0okWfbGJ3IWFFXRL0-70`

* Paso 2: generar la URL al archivo
`https://docs.google.com/spreadsheets/d/1FRLM9E7W4VVWEP_sx9QWq2w0okWfbGJ3IWFFXRL0-70/export?format=csv`

* Paso 3: si necesitas acceder a una hoja que no sea la primera, entonces a la URL agregas el identificador de la hoja con el & `&gid=1937180264 `

```python
# importar Pandas
import pandas as pd
```

```python
url = 'https://docs.google.com/spreadsheets/d/1FRLM9E7W4VVWEP_sx9QWq2w0okWfbGJ3IWFFXRL0-70/export?format=csv&gid=1937180264'
df = pd.read_csv(url)
```
